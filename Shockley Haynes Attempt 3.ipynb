```python
#importing packages I might use
import matplotlib.pyplot as plt ### plotting things
import numpy as np ## one of python's main maths packages
import pandas as pd ## for reading in our data
from scipy.optimize import curve_fit ## for fitting a line to our data
import matplotlib.ticker as ticker ## this one lets us change some parameters in our plots.
%matplotlib inline 
%config InlineBackend.figure_format = 'retina' 
from IPython.display import display, Markdown
import random
from itertools import cycle
from mpl_toolkits.mplot3d import Axes3D
from collections import Counter
```

```python
displacements = []
 


def walks():
    disp = 0
    steps = 10000 #number of time steps made, each time step can be for example 1ns
    chance = 0
    decay = 0 
    for x in range (0, steps):
        chance = np.random.uniform(0,1)
        decay = np.random.uniform(0,1)
        disp += 20 #drift velocity, note it is relative to the distance travelled in per step
        if decay < 0.01: #chance of decay
            break
        if chance < 0.5:
            disp += 1000 #distance travelled right in one step
        if chance > 0.5:
            disp -= 1000 #again here but left
    displacements.append(disp)
    

    
for _ in range(10000): #number here indicates number of walks ie number of electrons N at the start of simulation
    walks()

            


```

```python
#leave all this shit alone, it works
occurences = []
maxdisp = max(displacements)

mindisp = min(displacements)



for _ in range(mindisp,maxdisp+1):
    occurences.append(displacements.count(_))
    
maxn = max(occurences)    
 
    



```

```python

#again, pls leave alone
occurencesarray= np.array(occurences)
xaxis2 = np.arange(mindisp, maxdisp + 1)
print
displacementsarray = np.array(displacements)
fig = plt.figure(figsize=(9,8))
ax = fig.add_subplot(1,1,1)


ax.errorbar(xaxis2,           
             occurencesarray,              
              
                
             marker='.',             
             markersize = 8,
             markerfacecolor = 'blue',
             color='black',          
             linestyle='none',       
             capsize=6,              
             )

ax.set_xlabel('Displacement/ Units')
ax.set_ylabel('Number of Charge Carriers N')

ax.tick_params(direction='in',      # I like 'in', could be 'out' or both 'inout'
                length=7, 
               bottom='on',         # I want ticks on the bottom axes
                left='on',
                top='on',
                right='on',# A reasonable length
               )
plt.rcParams.update({'font.size':16})   # A decent font size so the text is readible. 
                                        # You may have to play with the aspect ration aobe and this to get a nice 
                                        # looking figure in your report.

plt.xlim(mindisp, maxdisp+1)                    # extent of the x axis, smallest to largest.
plt.ylim(0, maxn + 5)
plt.show()

```
